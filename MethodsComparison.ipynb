```python
def find_winners(input_array):
    rows, cols = input_array.shape
    winner_count = 0
    for col_num in range(cols):
        col = input_array[:, col_num]
        max_val = np.max(col)
        max_ind = np.argmax(col)

        if len(np.where(col == max_val)[0]) == 1 and max_ind == 0:
            winner_count += 1
    return winner_count
```
